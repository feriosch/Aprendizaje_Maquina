El proyecto tiene como objetivo construir un modelo de detección de fraudes en tarjetas de crédito. Usaremos la transacción y sus etiquetas como fraude o no fraude para detectar si las nuevas transacciones realizadas por el cliente son fraude o no.

* Análisis de sentimiento
    * El análisis de sentimientos es el proceso de analizar la emoción de los usuarios. Podemos categorizar sus emociones como positivas, negativas o neutrales. Es un gran proyecto para comprender cómo realizar análisis de sentimiento y se está utilizando ampliamente en la actualidad. Este es uno de los proyectos de aprendizaje automático más populares. La razón detrás de esto es que cada empresa está tratando de comprender el sentimiento de sus clientes si los clientes están contentos, se quedarán. Este proyecto podría mostrar un camino para reducir la rotación de clientes.
    
![img](https://d2h0cx97tjks2p.cloudfront.net/blogs/wp-content/uploads/sites/2/2020/01/sentiment-analysis-using-ML.jpg) 

* Reconocimiento de emociones del habla

    * El sistema de reconocimiento de emociones por voz utiliza datos de audio. Toma una parte del discurso como entrada y luego determina en qué emociones está hablando el hablante. Podemos identificar diferentes emociones como feliz, triste, sorprendido, enojado, etc. Este proyecto puede ser útil para identificar las emociones del cliente durante la llamada con el centro de llamadas.
